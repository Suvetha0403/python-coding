```python
data={'name':['Suve','jeni','sam','hari','bob'],'age':[16, 17, 45, 70, 25]}
data
```

```python
d=pd.DataFrame(data)
print(d)
```

```python
#-----------apply,---------------np.where,--------np.select
def categorize_age(row):
    if row['age'] <= 12:
        return 'Child'
    elif row['age'] <= 19:
        return 'Teen'
    elif row['age'] <= 64:
        return 'Adult'
    else:
        return 'Senior'
d['Age_grp'] = d.apply(categorize_age,axis=1)
print(d)
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df['range']= np.where(df.age>40,'middle','teen')
df
```

```python
df['range1']= np.where(df.age>40,'old',np.where(df.age>20,'middle','teen'))
df
```
